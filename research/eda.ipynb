```python
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data['Feature'] = df.columns
    vif_data['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif_data
```

```python
def reduce_vif(df, threshold=10.0):
    high_vif_features = []
    while True:
        vif_data = calculate_vif(df)
        max_vif = vif_data['VIF'].max()
        if max_vif > threshold:
            max_vif_feature = vif_data.loc[vif_data['VIF'] == max_vif, 'Feature'].values[0]
            high_vif_features.append(max_vif_feature)
            print(f'Removing {max_vif_feature} with VIF: {max_vif}')
            df = df.drop(columns=[max_vif_feature])
        else:
            break
    return df, high_vif_features

features = df.drop(columns=['target'])

# Reduce VIF
reduced_df, removed_features = reduce_vif(features)
print(removed_features)
```
